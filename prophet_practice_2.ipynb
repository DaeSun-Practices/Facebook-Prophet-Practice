```python
import pandas as pd
from fbprophet import Prophet
import pyupbit
```

```python
df = pyupbit.get_ohlcv("KRW-XRP", interval="minute1")

df = df.drop(["open", "high", "low", "volume", "value"], axis=1)

df = df.reset_index().rename(columns={"index": "ds"})
df = df.rename(columns={"close": "y"})
print(df)
```

```python
model = Prophet()
model.fit(df)
```
